```python
df = pd.read_csv("../input/salaries-test/Salaries.csv")
df
```

1) What is average of each rank (i.e.,Prof,AssocProf,and Asstprof)salary?

```python
avg_rank_salary = df.groupby(['rank'],sort = False)[['salary']].mean()
avg_rank_salary
```

2)What is the salary difference between male salary and female salary?

```python
sum_sex_salary = df.groupby(['sex'],sort = False)[['salary']].sum()
sum_male = sum_sex_salary['salary'][0]
sum_female = sum_sex_salary['salary'][1]
```

```python
diff_sex_salary = abs(sum_male-sum_female)
print('The salary difference between male salary and female salary is',diff_sex_salary)
```

3)What is the average salary of the male AsstProf?

```python
sort_AsstProf_Male = df[(df['rank'] == 'AsstProf')&(df['sex'] == 'Male')]
sort_AsstProf_Male
```

```python
avg_sort_AsstProf_Male = sort_AsstProf_Male.groupby(['rank','sex'],sort = False)[['salary']].mean()
avg_sort_AsstProf_Male
```

4)What is the average salary of the female Prof?

```python
sort_Prof_Female = df[(df['rank'] == 'Prof')&(df['sex'] == 'Female')]
```

```python
avg_sort_Prof_Female = sort_Prof_Female.groupby(['rank','sex'],sort = False)[['salary']].mean()
avg_sort_Prof_Female
```
